```python
from scipy.stats import pearsonr, spearmanr, ttest_ind
import pandas as pd
```

```python
all_srl = [
    'goal_setting', 
    'environment_structuring', 
    'task_strategies', 
    'time_management', 
    'help_seeking', 
    'self_evaluation', 
    'rehearsal', 
    'elaboration',
    'organization',
    'critical_thinking',
    'metacognitive_selfregulation',
    'math_self_efficacy',
    'internet_self_efficacy'
]
#df = pd.read_csv('eg2_avg.csv')
```

```python
df_cg = pd.read_csv('cg_avg.csv')
df_eg1 = pd.read_csv('eg1_avg.csv')
df_eg2 = pd.read_csv('eg2_avg.csv')
```

```python
data = {'name': [], 'stats': [], 'p-value': []}

for i in all_srl:
    result = ttest_ind(list(df_cg['a_'+i]), list(df_eg2['b_'+i]))
    
    data['name'].append(i)
    data['stats'].append(result[0])
    data['p-value'].append(result[1])
```

```python
df_result = pd.DataFrame(data)
# df_result.to_csv('cg_srl.csv')
df_result
```
