```python

def plot_wps (t):
    
    # constants for the signal
    A1 = 4.
    t1 = 1.
    f1 = 1.
    phi1 = 0.



    # generate the signal
    H1 = (np.sign(t - t1) + 1) / 2
    st1 = A1 * (t - t1) * np.exp(-2 * (t - t1))
    st1 *= np.cos(2. * np.pi * f1 * (t - t1) + phi1 * np.pi) * H1
    
    return st1



dt = 0.001
fmin = 15
fmax = 8000
```

```python
motor = 'Bearing1_1/'
path = '../datasets/phm-2012/Learning_set/' + motor
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
onlyfiles.sort()
```

```python
for file in onlyfiles:
    if int(file.split('_')[1].split('.')[0]) % 200 == 0:
        
        print(path+file)
        
        b11_pd = pd.read_csv(path+file, header=None)
        plot_tfr(plot_wps(b11_pd[4].values), dt=dt, fmin=fmin, fmax=fmax)
        
```
